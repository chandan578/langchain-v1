### Streaming And Batch

### Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses.
Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time:

```python
model.invoke("Write me 300 paragraph on Why men are so lazy??")
```

```python
model.stream("Write me 300 paragraph on Why men are so lazy??")  ## this create generator
```

```python
for chunk in model.stream("Write me 300 paragraph on Why men are so lazy??"):
    print(chunk.content, end=" ")
```

```python
for chunk in model.stream("Why do people are most attracted for sexual thing??"):
    print(chunk.content, end='', flush=True)
```

```python
for chunk in model.stream("How to avoid that form sexual thing??"):
    print(chunk.content, end='', flush=True)
```


#### Batch
Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel:

```python
response = model.batch([
    "Why parrots are so smart compared to other birds?",
    "Why parrots feather are so colorful??",
    "Why parrots can't fly long distances??",
    "How to parrot talk like a human??"
])


response
```

```python
model.batch([
    "Why parrots are so smart compared to other birds?",
    "Why parrots feather are so colorful??",
    "Why parrots can't fly long distances??",
    "How to parrot talk like a human??"],
    config={
        'max_concurrency': 5, ## limit the number of requests to 5
    }
)
```
